## IntroducciÃ³n

La sequÃ­a es un fenÃ³nemo meteorolÃ³gico en el cual se presentan significativamente menos precipitaciones pluviales de las normalmente esperadas. Dado que el agua es un recurso de vital importancia, su falta presenta un problema grave tanto para el adecuado balance de los ecosistemas como para el consumo humano.

Este fenÃ³meno es recurrente, aunque no cÃ­clico, y normalmente estÃ¡ regido por las fluctuaciones climÃ¡ticas, la actividad solar o la influencia de la temperatura ocÃ©anica [[ 1 ]](https://www.gob.mx/imta/articulos/perspectiva-sobre-la-sequia-actual-en-mexico?idiom=es) . Sin embargo, el cambio en los procesos climÃ¡ticos por parte humana ha provocado que este fenÃ³meno se profundice de manera extrema [[ 2 ]](https://www.ciad.mx/notas/item/2549-la-sequia-en-mexico-un-asunto-de-todos), causando problemas en la industria agropecuaria y ganadera, asÃ­ como generando problemas de acceso de agua potable y de uso diario para la poblaciÃ³n.

En MÃ©xico este problema es grande principalmente en el noroeste del paÃ­s, ya son regiones Ã¡ridas. De acuerdo con datos de la NASA, el paÃ­s se enfrenta a una de las sequÃ­as mÃ¡s intensas en los Ãºltimos aÃ±os, generando escasez de agua disponible para uso humano tanto individual como de producciÃ³n, puesto que las presas cuentan con niveles muy bajos de volumen, en concreto cerca de 60 presas se encuentran a menos del 25 por ciento de su capacidad [[ 3 ]](https://earthobservatory.nasa.gov/images/148270/widespread-drought-in-mexico), y mÃ¡s del 85 por ciento del territorio mexicano sufre de sequÃ­a [[ 4 ]](https://www.bbc.com/mundo/noticias-america-latina-57163233).

Dado que las precipitaciones pluviales estÃ¡n profundamente relacionadas con el cultivo, la sequÃ­a tiene un efecto grande en el sector agropecuario, pues la falta de agua genera que se obtenga menos producto de las cosechas [[ 5 ]](https://www.reuters.com/business/environment/mexico-water-supply-buckles-worsening-drought-putting-crops-risk-2021-07-02/).

Por lo tanto, este proyecto pretende realizar un anÃ¡lisis de los datos de la sequÃ­a meteorolÃ³gica en MÃ©xico, asÃ­ como su impacto en la agricultura. Con este fin se escogiÃ³ particularmente analizar las cosechas de maÃ­z, frijol y trigo; ya que el maÃ­z es la base de alimentaciÃ³n para gran parte de la poblaciÃ³n y representa un alimento de gran importancia para la poblaciÃ³n mexicana y para la economÃ­a del paÃ­s. En cuanto al frijol y al trigo, Ã©stos tambiÃ©n representan una parte importante de la dieta mexicana.

```python
def preparacion__de_scrapping(producto):
  options = webdriver.ChromeOptions()
  options.add_argument('--headless')
  options.add_argument('--no-sandbox')
  options.add_argument('--disable-dev-shm-usage')
  wd = webdriver.Chrome('chromedriver',options=options)
  wd.get("http://infosiap.siap.gob.mx:8080/agricola_siap_gobmx/ResumenProducto.do")

  select_moda = Select(wd.find_element_by_name('moda'))
  select_moda.select_by_index(1)

  select_producto = Select(wd.find_element_by_name('producto'))
  select_producto.select_by_visible_text(producto)

  select_ciclo = Select(wd.find_element_by_name('ciclo'))
  select_ciclo.select_by_index(0)

  radio_nivel = wd.find_elements_by_name('nivel')[0]
  radio_nivel.click()

  def scrapping(year_parametro, mes_parametro, indice, df_original):
    select_year = Select(wd.find_element_by_name('anio'))
    select_year.select_by_index(year_parametro)
    select_mes = Select(wd.find_element_by_name('mes'))
    select_mes.select_by_index(mes_parametro)

    wd.find_elements_by_xpath("//input[@value='Consultar']")[0].click()
    html = BeautifulSoup(wd.page_source, 'html.parser')
    table_html = html.find_all('table')[indice]

    df_aux = pd.read_html(str(table_html), flavor='bs4')[0]
    df_aux['AÃ±o'] = str(indice_year)
    df_aux['Mes'] = str(mes_parametro + 1)
    df_original = pd.concat([df_original, df_aux], ignore_index=True)
    return df_original

  df_scrapping = pd.DataFrame()
  indice_year = 2004

  for year in range(17):
    for mes in range(12):
      if mes != 11:
        try:
          df_scrapping = scrapping(year, mes, 2, df_scrapping)
        except:
          continue

      else:
        try:
          df_scrapping = scrapping(year, mes, 5, df_scrapping)
        except:
          continue
    indice_year += 1

  df_scrapping.columns = ['Estado', 'Sembrada_ha', 'Cosechada_ha', 'Siniestrada_ha', 'Siniestrada_ha', 'Rendimiento_ton_ha', 'AÃ±o', 'Mes']
  df_scrapping = df_scrapping.drop(['Siniestrada_ha', 'Siniestrada_ha', 'Rendimiento_ton_ha'], axis=1)
  df_scrapping = df_scrapping.dropna(subset=['Sembrada_ha', 'Cosechada_ha'], )
  df_scrapping = df_scrapping.reset_index(drop=True)
  df_scrapping['Estado'] = df_scrapping['Estado'].str.capitalize()
  df_scrapping['AÃ±o'] = df_scrapping['AÃ±o'].astype(np.int64)
  df_scrapping['Mes'] = df_scrapping['Mes'].astype(np.int64)

  return df_scrapping
```

El campo 'geometry' es esencialmente un vector donde cada entrada es un conjunto de formas correspondientes a una observaciÃ³n. Una entrada puede constar de una sola forma (como un Ãºnico polÃ­gono, Polygon) o de varias formas que deben considerarse como una sola observaciÃ³n (como los polÃ­gonos que componen las islas de un Estado, o excluyen los cuerpos de agua del suelo terrestre, MultiPolygon). En este caso es el estado de Veracruz. 

Para ello la funciÃ³n 'get_multipolygon_list()' devuelve la lista de coordenadas (x,y) del polÃ­gono (Polygon) o los polÃ­gonos que lo conforman (MultiPolygon)

```python
def get_multipolygon_list(geometria_poligonos_in: 'shapely...multiPolygon')->list:
  """
  Devuelve los poligonos en cada elemento Polygon en una lista
  """
  geometria_poligonos = geometria_poligonos_in.reset_index().loc[0,'geometry']
  list_ = []
  if (str(geometria_poligonos.type) == 'Polygon'):
    x, y = geometria_poligonos.exterior.coords.xy
    list_ = list(__builtin__.map(lambda x, y: [x, y], x, y))

  if (str(geometria_poligonos.type) == 'MultiPolygon'):
    for polygon_ in geometria_poligonos: 
      x, y = polygon_.exterior.coords.xy
      list_.append(list(zip(x, y)))
  return list_
```

Obtenemos de la API la pÃ©rdida forestal del estado de Veracruz:
- Primero obtenemos la lista de las coordenadas de los polÃ­gonos que lo conforman

```python
index_veracruz = 9 
cc_veracruz = get_multipolygon_list(entidades['geometry'][[int(f'{index_veracruz}')]])
cc_veracruz[0] # Primer coordenada de los poligonos
```

Utilizamos la manera anterior para acceder a los datos de pÃ©rdida forestal. AÃ±adimos que para la banda de pÃ©rdida anual nos retorne los datos de cada aÃ±o con addBands obtenidos de una agregaciÃ³n espacial reduceRegion en la geometrÃ­a del estado e indicamos los datos de escala y pixeles mÃ¡ximos a procesar.

```python
def get_loss_by_year(multi_polygon_list: list)->'copy_ee.dictionary.Dictionary':
  """
  Devuelve objeto copia de ee.dictionary.Dictionary con anidacion de dict.
  """
  gfc2020 = ee.Image('UMD/hansen/global_forest_change_2020_v1_8')
  imagen_perdida = gfc2020.select(['loss'])
  imagen_area_perdida = imagen_perdida.multiply(ee.Image.pixelArea()) 
  anios_perdida = gfc2020.select(['lossyear'])
  multi_polygon = ee.Geometry.MultiPolygon(ee.List(multi_polygon_list))
  perdida_anual = imagen_area_perdida.addBands(anios_perdida).reduceRegion(
    **{ 'reducer': ee.Reducer.sum().group(
        **{ 'groupField': 1}),
            'geometry': multi_polygon, 
            'scale': 30,
            'maxPixels': 1e9
      })

  return perdida_anual.getInfo().copy()
```

Obtenemos un diccionario anidado con el Ã­ndice y metros cuadrados de pÃ©rdida forestal

```python
perdida_veracruz = get_loss_by_year(cc_veracruz)
perdida_veracruz
```

Para explorar los datos usamos 'get_flatten_dict' que nos regresa una lista con el aÃ±o y perdida en metros cuadrados ambos en formato str.

```python
def get_flatten_dict(dict_in: 'obj_dict_copia')->'lista_perdida_por_anio_m2':
  """
  Devuelve una lista con la sublista del par anio-perdida en metros cuadrados.
  """
  return list(map(lambda dict_: [F"20{int(dict_.get('group')):02d}", dict_.get('sum')], dict_in.get('groups')))
```

En dado caso que el dataframe no este vacÃ­o se crea la figura utilizando la biblioteca seaborn y se retorna la figura.

```python
def plotting_siembre_cosecha(year, mes, producto):
  datos_productos = siembra_cosecha_sequia_del_maiz if producto =='Maiz' else (siembra_cosecha_sequia_del_frijol if producto == 'Frijol' else siembra_cosecha_sequia_del_trigo)
  datos_consulta = datos_productos.query(f'AÃ±o == {year} and Mes == {mes} and Estado != "Total"')
  datos_consulta = datos_consulta.melt(id_vars=['Estado', 'Tipo_sequia'],value_vars=['Sembrada_ha', 'Cosechada_ha'])

  if datos_consulta.empty:
    return "No hay datos para estas fechas, intente otra"
  else:
    g = sns.catplot(
      data=datos_consulta, kind="bar",
      x="Estado", y="value", hue="variable",
      ci=None, height=5, aspect=3
    )

    ax = g.facet_axis(0,0)
    for p in ax.patches:
        ax.text(p.get_x() + 0.10, 
                p.get_height() * 1.02, 
                '{0:.0f}'.format(p.get_height()), 
                color='black', rotation='horizontal', size='large', ha="left", va = 'bottom')

    g.despine(bottom=True)
    g.set(title="Sembrado y cosechado de "+ producto + " y el tipo de sequia que prensetaba en la fecha: " + str(year) + "-" + str(mes))
    g.set_axis_labels("Estado y Tipo de Sequia que presentaba","Superficie (ha)",fontsize = 20 )
    g.legend.set_title("Variable que se presenta")

    g.fig.set_figwidth(30)
    g.fig.set_figheight(10)

    return g
```

Para ver si hay una relaciÃ³n de la siembra y cosecha con el tipo de sequÃ­a debemos calcular tambiÃ©n si hubo muchas perdidas de hectÃ¡reas cuando existÃ­a sequÃ­a en el estado.

Hacemos una copia de los datos ya procesados de la siembra y cosecha de cada producto, procedemos a calcular la pÃ©rdida por hectÃ¡rea restando lo cosechado a lo sembrado, esto lo guardamos en una nueva columna llamada perdida_ha, hacemos el mismo procedimiento para cada dataframe.

```python
perdidas_maiz, perdidas_frijol, perdidas_trigo = siembra_cosecha_sequia_del_maiz.copy(), siembra_cosecha_sequia_del_frijol.copy(), siembra_cosecha_sequia_del_trigo.copy()
perdidas_maiz['perdida_ha'] = perdidas_maiz['Sembrada_ha'] - perdidas_maiz['Cosechada_ha']
perdidas_frijol['perdida_ha'] = perdidas_frijol['Sembrada_ha'] - perdidas_frijol['Cosechada_ha']
perdidas_trigo['perdida_ha'] = perdidas_trigo['Sembrada_ha'] - perdidas_trigo['Cosechada_ha']
```

Tenemos una lista de dataframes, y serÃ­a preferible tener un Ãºnico dataframe con todos los datos de todos los aÃ±os para el anÃ¡lisis estadÃ­stico. Para esto, hemos de acomodar nuestros datos de manera que los meses sean una variable "fecha" y estÃ©n relacionados a su promedio. De esta forma los anÃ¡lisis serÃ­an mÃ¡s sencillos a nivel lista, y en la concatenaciÃ³n todos los dataframes tendrÃ­an los mismos Ã­ndices de columnas

```python
def concatenar(dataframe, nombre): #funciÃ³n de concatenaciÃ³n de los dataframes
  def acomodar(data_frame, i, nombre): #funciÃ³n de acomodar los dataframes con el mes como variable
    return pd.melt(data_frame[i], id_vars= ["ENTIDAD"], value_vars= data_frame[i].columns[1:]).rename(columns = {"variable" : "FECHA", "value" : nombre})
  lista_acomodados = [] #lista de los dataframes acomodados
  for i in range(19): #aplicaciÃ³n de la funciÃ³n a los dataframes
   lista_acomodados.append(acomodar(dataframe, i, nombre)) #guardado de los dataframes a la lista
  dataframe_concatenados = pd.concat(lista_acomodados).reset_index(drop = True) #uniÃ³n de los dataframes
  return dataframe_concatenados
```
