## Introducción

La sequía es un fenónemo meteorológico en el cual se presentan significativamente menos precipitaciones pluviales de las normalmente esperadas. Dado que el agua es un recurso de vital importancia, su falta presenta un problema grave tanto para el adecuado balance de los ecosistemas como para el consumo humano.

Este fenómeno es recurrente, aunque no cíclico, y normalmente está regido por las fluctuaciones climáticas, la actividad solar o la influencia de la temperatura océanica [[ 1 ]](https://www.gob.mx/imta/articulos/perspectiva-sobre-la-sequia-actual-en-mexico?idiom=es) . Sin embargo, el cambio en los procesos climáticos por parte humana ha provocado que este fenómeno se profundice de manera extrema [[ 2 ]](https://www.ciad.mx/notas/item/2549-la-sequia-en-mexico-un-asunto-de-todos), causando problemas en la industria agropecuaria y ganadera, así como generando problemas de acceso de agua potable y de uso diario para la población.

En México este problema es grande principalmente en el noroeste del país, ya son regiones áridas. De acuerdo con datos de la NASA, el país se enfrenta a una de las sequías más intensas en los últimos años, generando escasez de agua disponible para uso humano tanto individual como de producción, puesto que las presas cuentan con niveles muy bajos de volumen, en concreto cerca de 60 presas se encuentran a menos del 25 por ciento de su capacidad [[ 3 ]](https://earthobservatory.nasa.gov/images/148270/widespread-drought-in-mexico), y más del 85 por ciento del territorio mexicano sufre de sequía [[ 4 ]](https://www.bbc.com/mundo/noticias-america-latina-57163233).

Dado que las precipitaciones pluviales están profundamente relacionadas con el cultivo, la sequía tiene un efecto grande en el sector agropecuario, pues la falta de agua genera que se obtenga menos producto de las cosechas [[ 5 ]](https://www.reuters.com/business/environment/mexico-water-supply-buckles-worsening-drought-putting-crops-risk-2021-07-02/).

Por lo tanto, este proyecto pretende realizar un análisis de los datos de la sequía meteorológica en México, así como su impacto en la agricultura. Con este fin se escogió particularmente analizar las cosechas de maíz, frijol y trigo; ya que el maíz es la base de alimentación para gran parte de la población y representa un alimento de gran importancia para la población mexicana y para la economía del país. En cuanto al frijol y al trigo, éstos también representan una parte importante de la dieta mexicana.

Dado que los datos de la siembra y cosecha están una página que no ofrece una API y sus datos solo se pueden obtener mediante archivos csv por año y mes, este proceso se haría muy tardado hacer lo manualmente, es por eso que se recurrió a usar web scraping para obtener la información ya que esta es presentada en forma de tabla, la siguiente función realiza el proceso de extracción de datos la [página](http://infosiap.siap.gob.mx:8080/agricola_siap_gobmx/ResumenProducto.do) de la SIAP.

```python
def preparacion__de_scrapping(producto):
  options = webdriver.ChromeOptions()
  options.add_argument('--headless')
  options.add_argument('--no-sandbox')
  options.add_argument('--disable-dev-shm-usage')
  wd = webdriver.Chrome('chromedriver',options=options)
  wd.get("http://infosiap.siap.gob.mx:8080/agricola_siap_gobmx/ResumenProducto.do")

  select_moda = Select(wd.find_element_by_name('moda'))
  select_moda.select_by_index(1)

  select_producto = Select(wd.find_element_by_name('producto'))
  select_producto.select_by_visible_text(producto)

  select_ciclo = Select(wd.find_element_by_name('ciclo'))
  select_ciclo.select_by_index(0)

  radio_nivel = wd.find_elements_by_name('nivel')[0]
  radio_nivel.click()

  def scrapping(year_parametro, mes_parametro, indice, df_original):
    select_year = Select(wd.find_element_by_name('anio'))
    select_year.select_by_index(year_parametro)
    select_mes = Select(wd.find_element_by_name('mes'))
    select_mes.select_by_index(mes_parametro)

    wd.find_elements_by_xpath("//input[@value='Consultar']")[0].click()
    html = BeautifulSoup(wd.page_source, 'html.parser')
    table_html = html.find_all('table')[indice]

    df_aux = pd.read_html(str(table_html), flavor='bs4')[0]
    df_aux['Año'] = str(indice_year)
    df_aux['Mes'] = str(mes_parametro + 1)
    df_original = pd.concat([df_original, df_aux], ignore_index=True)
    return df_original

  df_scrapping = pd.DataFrame()
  indice_year = 2004

  for year in range(17):
    for mes in range(12):
      if mes != 11:
        try:
          df_scrapping = scrapping(year, mes, 2, df_scrapping)
        except:
          continue

      else:
        try:
          df_scrapping = scrapping(year, mes, 5, df_scrapping)
        except:
          continue
    indice_year += 1

  df_scrapping.columns = ['Estado', 'Sembrada_ha', 'Cosechada_ha', 'Siniestrada_ha', 'Siniestrada_ha', 'Rendimiento_ton_ha', 'Año', 'Mes']
  df_scrapping = df_scrapping.drop(['Siniestrada_ha', 'Siniestrada_ha', 'Rendimiento_ton_ha'], axis=1)
  df_scrapping = df_scrapping.dropna(subset=['Sembrada_ha', 'Cosechada_ha'], )
  df_scrapping = df_scrapping.reset_index(drop=True)
  df_scrapping['Estado'] = df_scrapping['Estado'].str.capitalize()
  df_scrapping['Año'] = df_scrapping['Año'].astype(np.int64)
  df_scrapping['Mes'] = df_scrapping['Mes'].astype(np.int64)

  return df_scrapping
```

El campo 'geometry' es esencialmente un vector donde cada entrada es un conjunto de formas correspondientes a una observación. Una entrada puede constar de una sola forma (como un único polígono, Polygon) o de varias formas que deben considerarse como una sola observación (como los polígonos que componen las islas de un Estado, o excluyen los cuerpos de agua del suelo terrestre, MultiPolygon). En este caso es el estado de Veracruz. 

Para ello la función 'get_multipolygon_list()' devuelve la lista de coordenadas (x,y) del polígono (Polygon) o los polígonos que lo conforman (MultiPolygon)

```python
def get_multipolygon_list(geometria_poligonos_in: 'shapely...multiPolygon')->list:
  """
  Devuelve los poligonos en cada elemento Polygon en una lista
  """
  geometria_poligonos = geometria_poligonos_in.reset_index().loc[0,'geometry']
  list_ = []
  if (str(geometria_poligonos.type) == 'Polygon'):
    x, y = geometria_poligonos.exterior.coords.xy
    list_ = list(__builtin__.map(lambda x, y: [x, y], x, y))

  if (str(geometria_poligonos.type) == 'MultiPolygon'):
    for polygon_ in geometria_poligonos: 
      x, y = polygon_.exterior.coords.xy
      list_.append(list(zip(x, y)))
  return list_
```

###### Pérdida Forestal

Obtenemos de la API la pérdida forestal del estado de Veracruz:
- Primero obtenemos la lista de las coordenadas de los polígonos que lo conforman

```python
index_veracruz = 9 
cc_veracruz = get_multipolygon_list(entidades['geometry'][[int(f'{index_veracruz}')]])
cc_veracruz[0] # Primer coordenada de los poligonos
```

Utilizamos la manera anterior para acceder a los datos de pérdida forestal. Añadimos que para la banda de pérdida anual nos retorne los datos de cada año con addBands obtenidos de una agregación espacial reduceRegion en la geometría del estado e indicamos los datos de escala y pixeles máximos a procesar.

```python
def get_loss_by_year(multi_polygon_list: list)->'copy_ee.dictionary.Dictionary':
  """
  Devuelve objeto copia de ee.dictionary.Dictionary con anidacion de dict.
  """
  gfc2020 = ee.Image('UMD/hansen/global_forest_change_2020_v1_8')
  imagen_perdida = gfc2020.select(['loss'])
  imagen_area_perdida = imagen_perdida.multiply(ee.Image.pixelArea()) 
  anios_perdida = gfc2020.select(['lossyear'])
  multi_polygon = ee.Geometry.MultiPolygon(ee.List(multi_polygon_list))
  perdida_anual = imagen_area_perdida.addBands(anios_perdida).reduceRegion(
    **{ 'reducer': ee.Reducer.sum().group(
        **{ 'groupField': 1}),
            'geometry': multi_polygon, 
            'scale': 30,
            'maxPixels': 1e9
      })

  return perdida_anual.getInfo().copy()
```

Obtenemos un diccionario anidado con el índice y metros cuadrados de pérdida forestal

```python
perdida_veracruz = get_loss_by_year(cc_veracruz)
perdida_veracruz
```

Para explorar los datos usamos 'get_flatten_dict' que nos regresa una lista con el año y perdida en metros cuadrados ambos en formato str.

```python
def get_flatten_dict(dict_in: 'obj_dict_copia')->'lista_perdida_por_anio_m2':
  """
  Devuelve una lista con la sublista del par anio-perdida en metros cuadrados.
  """
  return list(map(lambda dict_: [F"20{int(dict_.get('group')):02d}", dict_.get('sum')], dict_in.get('groups')))
```

Vamos a utlizar un formulario de colab para presentar los datos de la siembra y cosecha, donde se podrá elegir el tipo de producto(maíz, frijol y trigo), el año y mes, para ello es conveniente crear una función para automatizar el proceso.

La función recibe como parametro el año, mes y el tipo de producto seleccionado, primero seleccinamos uno de los tres dataframes, esto se hace dependiendo del valor de la variable producto, después se filtra dependiendo de los valores seleccionados, nótese que se filtra que estado no sea igual a Total ya que los datos contienen esa variable por año.

Proseguimos creando un nuevo dataframe usando la función melt. Si el nuevo dataframe esta vacío quiere decir que no hay datos para esa fecha y se regresa un texto.
En dado caso que el dataframe no este vacío se crea la figura utilizando la biblioteca seaborn y se retorna la figura.

```python
def plotting_siembre_cosecha(year, mes, producto):
  datos_productos = siembra_cosecha_sequia_del_maiz if producto =='Maiz' else (siembra_cosecha_sequia_del_frijol if producto == 'Frijol' else siembra_cosecha_sequia_del_trigo)
  datos_consulta = datos_productos.query(f'Año == {year} and Mes == {mes} and Estado != "Total"')
  datos_consulta = datos_consulta.melt(id_vars=['Estado', 'Tipo_sequia'],value_vars=['Sembrada_ha', 'Cosechada_ha'])

  if datos_consulta.empty:
    return "No hay datos para estas fechas, intente otra"
  else:
    g = sns.catplot(
      data=datos_consulta, kind="bar",
      x="Estado", y="value", hue="variable",
      ci=None, height=5, aspect=3
    )

    ax = g.facet_axis(0,0)
    for p in ax.patches:
        ax.text(p.get_x() + 0.10, 
                p.get_height() * 1.02, 
                '{0:.0f}'.format(p.get_height()), 
                color='black', rotation='horizontal', size='large', ha="left", va = 'bottom')

    g.despine(bottom=True)
    g.set(title="Sembrado y cosechado de "+ producto + " y el tipo de sequia que prensetaba en la fecha: " + str(year) + "-" + str(mes))
    g.set_axis_labels("Estado y Tipo de Sequia que presentaba","Superficie (ha)",fontsize = 20 )
    g.legend.set_title("Variable que se presenta")

    g.fig.set_figwidth(30)
    g.fig.set_figheight(10)

    return g
```

##### Datos sobre la temperatura de México (CONAGUA)

Tenemos una lista de dataframes, y sería preferible tener un único dataframe con todos los datos de todos los años para el análisis estadístico. Para esto, hemos de acomodar nuestros datos de manera que los meses sean una variable "fecha" y estén relacionados a su promedio. De esta forma los análisis serían más sencillos a nivel lista, y en la concatenación todos los dataframes tendrían los mismos índices de columnas

```python
def concatenar(dataframe, nombre): #función de concatenación de los dataframes
  def acomodar(data_frame, i, nombre): #función de acomodar los dataframes con el mes como variable
    return pd.melt(data_frame[i], id_vars= ["ENTIDAD"], value_vars= data_frame[i].columns[1:]).rename(columns = {"variable" : "FECHA", "value" : nombre})
  lista_acomodados = [] #lista de los dataframes acomodados
  for i in range(19): #aplicación de la función a los dataframes
   lista_acomodados.append(acomodar(dataframe, i, nombre)) #guardado de los dataframes a la lista
  dataframe_concatenados = pd.concat(lista_acomodados).reset_index(drop = True) #unión de los dataframes
  return dataframe_concatenados
```

```python
def separacion_de_fecha(dataframe): #función que separa la columna fecha
  ultima_columna = dataframe.columns[2]
  dataframe[["Año", "Mes"]] = dataframe["FECHA"].str.split("-", expand = True) #separación de la columna
  dataframe = dataframe.drop(columns = "FECHA") #eliminación de la columna original
  dataframe = dataframe[["ENTIDAD", "Año", "Mes", ultima_columna]] #ordenamiento de las columnas
  return dataframe 
```

## Conclusiones

### ¿Cómo ha evolucionado la sequía en el país?

Sin duda alguna el año 2011 es donde más sequías graves se han presentado en México, sin embargo, desde el año 2019 hasta la fecha más municipios han presentado sequías, incluso en el año 2020 por la pandemia que surgió en el mundo debido al virus SARS-CoV-2 la demanda de agua en el país aumento y con ellos muchos municipios pasaron a tener tipos de sequía D2 y D3. En resumidas cuentas, México está pasando por una sequía preocupante en el año actual 2021. Los resultados de la evolución se pueden ver en el heat map y los mapas sobre la sequía.

### ¿Cuáles son los estados o regiones con mayor problema de sequía?

Principalmente los estados del norte del país son los que más sequias han presentado y son los únicos que han presentado una sequía tipo D4, que es la categoría más grave puesta por la CONAGUA, estos estados son Sonora, Chihuahua, Coahuila, Tamaulipas, Sinaloa y Durango.

### ¿En que meses hay mas tiempo de sequía?

El mes donde más sequia se ha presentado es el mes de junio, seguido de mayo y junio, estos meses son los que con más frecuencia la sequía aumenta, cabe destacar que junio por lo general siempre es el mes con mayores temperaturas, en los estados mencionados en la pregunta pasada.

### ¿Existe un impacto entre la sequía y producción de maíz, frijol y trigo en México? 

Si existe un impacto, pero también la sequía no es un factor que afecte directamente en lo sembrado y cosechado, vimos estados que presentaban sequias y aun así no tuvieron muchas pérdidas, sin embargo otros estados que si presentaban sequías tuvieron pérdidas, podemos decir que la sequía es uno de los factores que puede influir en la producción de estos tres alimentos.

Aproximadamente el 82% de los casos por año cumplen las condiciones para la siembra de maíz, estamos hablando de aproximadamente 314 en promedio de 384 casos por año

En una primera aproximación se podría decir que no, de hecho pareciera ocurrir todo lo contrario, ya que desde el 2012, el número de casos optimos no ha bajado de 320 por año, muy posiblemente se debe a que México en cuestión de casos no óptimos, tiene problemas con las temperaturas bajas, entonces el calentamiento global podría estar ocasionando que esos casos entren en el rango optimo, aunque estas son meras especulaciones.
