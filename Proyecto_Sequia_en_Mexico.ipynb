## Introducci√≥n

La sequ√≠a es un fen√≥nemo meteorol√≥gico en el cual se presentan significativamente menos precipitaciones pluviales de las normalmente esperadas. Dado que el agua es un recurso de vital importancia, su falta presenta un problema grave tanto para el adecuado balance de los ecosistemas como para el consumo humano.

Este fen√≥meno es recurrente, aunque no c√≠clico, y normalmente est√° regido por las fluctuaciones clim√°ticas, la actividad solar o la influencia de la temperatura oc√©anica [[ 1 ]](https://www.gob.mx/imta/articulos/perspectiva-sobre-la-sequia-actual-en-mexico?idiom=es) . Sin embargo, el cambio en los procesos clim√°ticos por parte humana ha provocado que este fen√≥meno se profundice de manera extrema [[ 2 ]](https://www.ciad.mx/notas/item/2549-la-sequia-en-mexico-un-asunto-de-todos), causando problemas en la industria agropecuaria y ganadera, as√≠ como generando problemas de acceso de agua potable y de uso diario para la poblaci√≥n.

En M√©xico este problema es grande principalmente en el noroeste del pa√≠s, ya son regiones √°ridas. De acuerdo con datos de la NASA, el pa√≠s se enfrenta a una de las sequ√≠as m√°s intensas en los √∫ltimos a√±os, generando escasez de agua disponible para uso humano tanto individual como de producci√≥n, puesto que las presas cuentan con niveles muy bajos de volumen, en concreto cerca de 60 presas se encuentran a menos del 25 por ciento de su capacidad [[ 3 ]](https://earthobservatory.nasa.gov/images/148270/widespread-drought-in-mexico), y m√°s del 85 por ciento del territorio mexicano sufre de sequ√≠a [[ 4 ]](https://www.bbc.com/mundo/noticias-america-latina-57163233).

Dado que las precipitaciones pluviales est√°n profundamente relacionadas con el cultivo, la sequ√≠a tiene un efecto grande en el sector agropecuario, pues la falta de agua genera que se obtenga menos producto de las cosechas [[ 5 ]](https://www.reuters.com/business/environment/mexico-water-supply-buckles-worsening-drought-putting-crops-risk-2021-07-02/).

Por lo tanto, este proyecto pretende realizar un an√°lisis de los datos de la sequ√≠a meteorol√≥gica en M√©xico, as√≠ como su impacto en la agricultura. Con este fin se escogi√≥ particularmente analizar la cosecha de ma√≠z, ya que es la base de alimentaci√≥n para gran parte de la poblaci√≥n y representa un alimento de gran importancia para la poblaci√≥n mexicana y para la econom√≠a del pa√≠s.

##### Datos sobre siembra y cosecha de ma√≠z, frijol y trigo en M√©xico (SIAP)

Dado que los datos de la siembre y cosecha est√°n una p√°gina que no ofrece una API y sus datos solo se pueden obtener mediante archivos csv por a√±o y mes, este proceso se har√≠a muy tardado hacer lo manualmente, es por eso que se recurri√≥ a usar web scraping para obtener la informaci√≥n ya que esta es presentada en forma de tabla, la siguiente funci√≥n realiza el proceso de extracci√≥n de datos la [p√°gina](http://infosiap.siap.gob.mx:8080/agricola_siap_gobmx/ResumenProducto.do) de la SIAP.

```python
def preparacion__de_scrapping(producto):
  options = webdriver.ChromeOptions()
  options.add_argument('--headless')
  options.add_argument('--no-sandbox')
  options.add_argument('--disable-dev-shm-usage')
  wd = webdriver.Chrome('chromedriver',options=options)
  wd.get("http://infosiap.siap.gob.mx:8080/agricola_siap_gobmx/ResumenProducto.do")

  select_moda = Select(wd.find_element_by_name('moda'))
  select_moda.select_by_index(1)

  select_producto = Select(wd.find_element_by_name('producto'))
  select_producto.select_by_visible_text(producto)

  select_ciclo = Select(wd.find_element_by_name('ciclo'))
  select_ciclo.select_by_index(0)

  radio_nivel = wd.find_elements_by_name('nivel')[0]
  radio_nivel.click()

  def scrapping(year_parametro, mes_parametro, indice, df_original):
    select_year = Select(wd.find_element_by_name('anio'))
    select_year.select_by_index(year_parametro)
    select_mes = Select(wd.find_element_by_name('mes'))
    select_mes.select_by_index(mes_parametro)

    wd.find_elements_by_xpath("//input[@value='Consultar']")[0].click()
    html = BeautifulSoup(wd.page_source, 'html.parser')
    table_html = html.find_all('table')[indice]

    df_aux = pd.read_html(str(table_html), flavor='bs4')[0]
    df_aux['A√±o'] = str(indice_year)
    df_aux['Mes'] = str(mes_parametro + 1)
    df_original = pd.concat([df_original, df_aux], ignore_index=True)
    return df_original

  df_scrapping = pd.DataFrame()
  indice_year = 2004

  for year in range(17):
    for mes in range(12):
      if mes != 11:
        try:
          df_scrapping = scrapping(year, mes, 2, df_scrapping)
        except:
          continue

      else:
        try:
          df_scrapping = scrapping(year, mes, 5, df_scrapping)
        except:
          continue
    indice_year += 1

  df_scrapping.columns = ['Estado', 'Sembrada_ha', 'Cosechada_ha', 'Siniestrada_ha', 'Siniestrada_ha', 'Rendimiento_ton_ha', 'A√±o', 'Mes']
  df_scrapping = df_scrapping.drop(['Siniestrada_ha', 'Siniestrada_ha', 'Rendimiento_ton_ha'], axis=1)
  df_scrapping = df_scrapping.dropna(subset=['Sembrada_ha', 'Cosechada_ha'], )
  df_scrapping = df_scrapping.reset_index(drop=True)
  df_scrapping['Estado'] = df_scrapping['Estado'].str.capitalize()
  df_scrapping['A√±o'] = df_scrapping['A√±o'].astype(np.int64)
  df_scrapping['Mes'] = df_scrapping['Mes'].astype(np.int64)

  return df_scrapping
```

```python
def formato(dataframe): #funci√≥n que cambia el formato por lista
  def numero(data_frame): #funci√≥n que cambia el formato por dataframe
   for col in data_frame.columns[1:]: #asignaci√≥n de las columnas a cambiar
     data_frame[col] = pd.to_numeric(data_frame[col]) #cambio del formato
  for dfs in range((len(dataframe)+1)): #asignaci√≥n de los dataframes a cambiar
   try:
     numero(dataframe[dfs]) #aplicaci√≥n de la funci√≥n de cambio de formato a los dataframes
   except:
     continue
```

Vamos a utlizar un formulario de colab para presentar los datos de la siembre y cosecha, donde se podra elegir el tipo de producto(ma√≠z, frijol y trigo), el a√±o y mes, para ello es conveniente crear una funcion para autimatizar el proceso.

La funci√≥n recibe como parametro el a√±o, mes y el tipo de producto seleccionado, primero seleccinamos uno de los tres dataframes, esto se hace dependiendo del valor de la variable producto, despu√©s se filtra dependiendo de los valores seleccionados, n√≥tese que se filtra que estado no sea igual a Total ya que los datos contienen esa variable por a√±o.

Proseguimos creando un nuevo dataframe usando la funci√≥n melt. Si el nuevo dataframe esta vacio quiere decir que no hay datos para esa fecha y se regres un texto.
En dado caso que el dataframe no este vacio se crea la figura utilizando la biblioteca seaborn y se retorna la figura.

```python
def plotting_siembre_cosecha(year, mes, producto):
  datos_productos = siembra_cosecha_sequia_del_maiz if producto =='Maiz' else (siembra_cosecha_sequia_del_frijol if producto == 'Frijol' else siembra_cosecha_sequia_del_trigo)
  datos_consulta = datos_productos.query(f'A√±o == {year} and Mes == {mes} and Estado != "Total"')
  datos_consulta = datos_consulta.melt(id_vars=['Estado', 'Tipo_sequia'],value_vars=['Sembrada_ha', 'Cosechada_ha'])

  if datos_consulta.empty:
    return "No hay datos para estas fechas, intente otra"
  else:
    g = sns.catplot(
      data=datos_consulta, kind="bar",
      x="Estado", y="value", hue="variable",
      ci=None, height=5, aspect=3
    )

    ax = g.facet_axis(0,0)
    for p in ax.patches:
        ax.text(p.get_x() + 0.10, 
                p.get_height() * 1.02, 
                '{0:.0f}'.format(p.get_height()), 
                color='black', rotation='horizontal', size='large', ha="left", va = 'bottom')

    g.despine(bottom=True)
    g.set(title="Sembrado y cosechado de "+ producto + " y el tipo de sequia que prensetaba en la fecha: " + str(year) + "-" + str(mes))
    g.set_axis_labels("Estado y Tipo de Sequia que presentaba","Superficie (ha)",fontsize = 20 )
    g.legend.set_title("Variable que se presenta")

    g.fig.set_figwidth(30)
    g.fig.set_figheight(10)

    return g
```

Para ver si hay una relaci√≥n de la siembra y cosecha con el tipo de sequ√≠a debemos calcular tambi√©n si hubo muchas perdidas de hect√°reas cuando exist√≠a sequ√≠a en el estado.

Hacemos una copia de los datos ya procesados de la siembre y cosecha de cada producto, procedemos a calcular la perdida por hect√°rea restando le lo cosechado a lo sembrado, esto lo guardamos en una nueva columna llamada perdida_ha, hacemos el mismo procedimiento para cada dataframe.

```python
perdidas_maiz, perdidas_frijol, perdidas_trigo = siembra_cosecha_sequia_del_maiz.copy(), siembra_cosecha_sequia_del_frijol.copy(), siembra_cosecha_sequia_del_trigo.copy()
perdidas_maiz['perdida_ha'] = perdidas_maiz['Sembrada_ha'] - perdidas_maiz['Cosechada_ha']
perdidas_frijol['perdida_ha'] = perdidas_frijol['Sembrada_ha'] - perdidas_frijol['Cosechada_ha']
perdidas_trigo['perdida_ha'] = perdidas_trigo['Sembrada_ha'] - perdidas_trigo['Cosechada_ha']
```

```python
productos_full = pd.concat([perdidas_maiz, perdidas_frijol, perdidas_trigo])
productos_full.head(2)
```

La siguiente funcion crea la grafica de perdidas por producto, a√±o y mes.

Recibe como parametro el a√±o y mes que se solicita, hace una consulta a dataframe concatenado, si esta consuta esta vac√≠a regresa un mensaje notificando que no hay datos para la fecha solicitada.
En dado caso que si existan datos creamos una grafica de barras para cada producto.


```python
def perdidas_por_producto(year, mes):
  perdidas_fecha = productos_full.query(f'A√±o == {year} and Mes == {mes}')
  if perdidas_fecha.empty:
    return "No hay datos para estas fechas, intente otra"
  else:
    g = sns.FacetGrid(perdidas_fecha, row="producto", height=5, aspect=3)
    g.map_dataframe(sns.barplot, "Estado", "perdida_ha")

    for i in range(3):
      ax = g.facet_axis(i,0)
      for p in ax.patches:
          ax.text(p.get_x() + 0.15, 
                  p.get_height() * 1.02, 
                  '{0:.0f}'.format(p.get_height()), 
                  color='black', rotation='horizontal', size='large', ha="left", va = 'bottom')
          
    g.set_axis_labels("Estado y Tipo de Sequia que presentaba","Superficie(ha)",fontsize = 20 )

    g.fig.set_figwidth(40)
    g.fig.set_figheight(10)

    return g
```

Pero al final tenemos una lista de dataframes, ser√≠a preferible tener un √∫nico dataframe con todos los datos de todos los a√±os para el an√°lisis estad√≠stico. Para esto, hemos de acomodar nuestros datos de manera que los meses sean una variable "fecha" y est√©n relacionados a su promedio. De esta forma los an√°lisis ser√≠an m√°s sencillos a nivel lista, y en la concatenaci√≥n todos los dataframes tendr√≠an los mismos √≠ndices de columnas

```python
def concatenar(dataframe, nombre): #funci√≥n de concatenaci√≥n de los dataframes
  def acomodar(data_frame, i, nombre): #funci√≥n de acomodar los dataframes con el mes como variable
    return pd.melt(data_frame[i], id_vars= ["ENTIDAD"], value_vars= data_frame[i].columns[1:]).rename(columns = {"variable" : "FECHA", "value" : nombre})
  lista_acomodados = [] #lista de los dataframes acomodados
  for i in range(19): #aplicaci√≥n de la funci√≥n a los dataframes
   lista_acomodados.append(acomodar(dataframe, i, nombre)) #guardado de los dataframes a la lista
  dataframe_concatenados = pd.concat(lista_acomodados).reset_index(drop = True) #uni√≥n de los dataframes
  return dataframe_concatenados
```

```python
Temp_promedio = concatenar(Temp_promedio, "Temperatura_promedio") #aplicamos la funci√≥n de concatenaci√≥n
```

Nos gustar√≠a que la columna "Fecha" estuviera dividida por "A√±o" y "Mes", para que sean m√°s f√°ciles las agrupaciones

```python
def separacion_de_fecha(dataframe): #funci√≥n que separa la columna fecha
  ultima_columna = dataframe.columns[2]
  dataframe[["A√±o", "Mes"]] = dataframe["FECHA"].str.split("-", expand = True) #separaci√≥n de la columna
  dataframe = dataframe.drop(columns = "FECHA") #eliminaci√≥n de la columna original
  dataframe = dataframe[["ENTIDAD", "A√±o", "Mes", ultima_columna]] #ordenamiento de las columnas
  return dataframe 
```
