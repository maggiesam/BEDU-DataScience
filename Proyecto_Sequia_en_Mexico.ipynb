## Introducción

La sequía es un fenónemo meteorológico en el cual se presentan significativamente menos precipitaciones pluviales de las normalmente esperadas. Dado que el agua es un recurso de vital importancia, su falta presenta un problema grave tanto para el adecuado balance de los ecosistemas como para el consumo humano.

Este fenómeno es recurrente, aunque no cíclico, y normalmente está regido por las fluctuaciones climáticas, la actividad solar o la influencia de la temperatura océanica [[ 1 ]](https://www.gob.mx/imta/articulos/perspectiva-sobre-la-sequia-actual-en-mexico?idiom=es) . Sin embargo, el cambio en los procesos climáticos por parte humana ha provocado que este fenómeno se profundice de manera extrema [[ 2 ]](https://www.ciad.mx/notas/item/2549-la-sequia-en-mexico-un-asunto-de-todos), causando problemas en la industria agropecuaria y ganadera, así como generando problemas de acceso de agua potable y de uso diario para la población.

En México este problema es grande principalmente en el noroeste del país, ya son regiones áridas. De acuerdo con datos de la NASA, el país se enfrenta a una de las sequías más intensas en los últimos años, generando escasez de agua disponible para uso humano tanto individual como de producción, puesto que las presas cuentan con niveles muy bajos de volumen, en concreto cerca de 60 presas se encuentran a menos del 25 por ciento de su capacidad [[ 3 ]](https://earthobservatory.nasa.gov/images/148270/widespread-drought-in-mexico), y más del 85 por ciento del territorio mexicano sufre de sequía [[ 4 ]](https://www.bbc.com/mundo/noticias-america-latina-57163233).

Dado que las precipitaciones pluviales están profundamente relacionadas con el cultivo, la sequía tiene un efecto grande en el sector agropecuario, pues la falta de agua genera que se obtenga menos producto de las cosechas [[ 5 ]](https://www.reuters.com/business/environment/mexico-water-supply-buckles-worsening-drought-putting-crops-risk-2021-07-02/).

Por lo tanto, este proyecto pretende realizar un análisis de los datos de la sequía meteorológica en México, así como su impacto en la agricultura. Con este fin se escogió particularmente analizar la cosecha de maíz, ya que es la base de alimentación para gran parte de la población y representa un alimento de gran importancia para la población mexicana y para la economía del país.

##### Datos sobre siembra y cosecha de maíz, frijol y trigo en México (SIAP)

Dado que los datos de la siembre y cosecha están una página que no ofrece una API y sus datos solo se pueden obtener mediante archivos csv por año y mes, este proceso se haría muy tardado hacer lo manualmente, es por eso que se recurrió a usar web scraping para obtener la información ya que esta es presentada en forma de tabla, la siguiente función realiza el proceso de extracción de datos la [página](http://infosiap.siap.gob.mx:8080/agricola_siap_gobmx/ResumenProducto.do) de la SIAP.

```python
def preparacion__de_scrapping(producto):
  options = webdriver.ChromeOptions()
  options.add_argument('--headless')
  options.add_argument('--no-sandbox')
  options.add_argument('--disable-dev-shm-usage')
  wd = webdriver.Chrome('chromedriver',options=options)
  wd.get("http://infosiap.siap.gob.mx:8080/agricola_siap_gobmx/ResumenProducto.do")

  select_moda = Select(wd.find_element_by_name('moda'))
  select_moda.select_by_index(1)

  select_producto = Select(wd.find_element_by_name('producto'))
  select_producto.select_by_visible_text(producto)

  select_ciclo = Select(wd.find_element_by_name('ciclo'))
  select_ciclo.select_by_index(0)

  radio_nivel = wd.find_elements_by_name('nivel')[0]
  radio_nivel.click()

  def scrapping(year_parametro, mes_parametro, indice, df_original):
    select_year = Select(wd.find_element_by_name('anio'))
    select_year.select_by_index(year_parametro)
    select_mes = Select(wd.find_element_by_name('mes'))
    select_mes.select_by_index(mes_parametro)

    wd.find_elements_by_xpath("//input[@value='Consultar']")[0].click()
    html = BeautifulSoup(wd.page_source, 'html.parser')
    table_html = html.find_all('table')[indice]

    df_aux = pd.read_html(str(table_html), flavor='bs4')[0]
    df_aux['Año'] = str(indice_year)
    df_aux['Mes'] = str(mes_parametro + 1)
    df_original = pd.concat([df_original, df_aux], ignore_index=True)
    return df_original

  df_scrapping = pd.DataFrame()
  indice_year = 2004

  for year in range(17):
    for mes in range(12):
      if mes != 11:
        try:
          df_scrapping = scrapping(year, mes, 2, df_scrapping)
        except:
          continue

      else:
        try:
          df_scrapping = scrapping(year, mes, 5, df_scrapping)
        except:
          continue
    indice_year += 1

  df_scrapping.columns = ['Estado', 'Sembrada_ha', 'Cosechada_ha', 'Siniestrada_ha', 'Siniestrada_ha', 'Rendimiento_ton_ha', 'Año', 'Mes']
  df_scrapping = df_scrapping.drop(['Siniestrada_ha', 'Siniestrada_ha', 'Rendimiento_ton_ha'], axis=1)
  df_scrapping = df_scrapping.dropna(subset=['Sembrada_ha', 'Cosechada_ha'], )
  df_scrapping = df_scrapping.reset_index(drop=True)
  df_scrapping['Estado'] = df_scrapping['Estado'].str.capitalize()
  df_scrapping['Año'] = df_scrapping['Año'].astype(np.int64)
  df_scrapping['Mes'] = df_scrapping['Mes'].astype(np.int64)

  return df_scrapping
```

Las fechas del registró estan como nombre de columna pero en formato fecha y hora, pero nos interesa que solo aparezca la fecha. Para renombramos las columnas que tenga ese formato para que aparezca en formato yyyy-m-d

```python
lista_fechas = list(datos_sequia_mexico.columns[9:]) # Seleccionamos el nombre de las columnas que vamos a modificar
lista_fechas_corregido = [i[:10] for i in lista_fechas] # Seleccionamos solo la fecha de los nombres
diccionario_de_fechas = {key:value for key, value in zip(lista_fechas, lista_fechas_corregido)} # Creamos un diccionario con las fechas anteriores y las nuevas
datos_sequia_mexico = datos_sequia_mexico.rename(columns=diccionario_de_fechas) # Renombramos las columnas
```

Como vemos todas las columnas que tienen de nombre una fecha tienen datos faltantes, los NaN que aparecen significa que en esa fecha ese municipio no presento sequía. Sin embargo cambiaremos los Nan por las siglas NPS(No Presenta Sequía).

La simboligía de los tipos de sequia es la siguiente:

*   **NPS:** No presenta Sequía.
*   **D0:** Anormalmente seco.
*   **D1:** Sequía moderada.
*   **D2:** Sequía severa.
*   **D3:** Sequía extrema.
*   **D4:** Sequía excepcional.

Vamos a utlizar un formulario de colab para presentar los datos de la siembre y cosecha, donde se podra elegir el tipo de producto(maíz, frijol y trigo), el año y mes, para ello es conveniente crear una funcion para autimatizar el proceso.

La función recibe como parametro el año, mes y el tipo de producto seleccionado, primero seleccinamos uno de los tres dataframes, esto se hace dependiendo del valor de la variable producto, después se filtra dependiendo de los valores seleccionados, nótese que se filtra que estado no sea igual a Total ya que los datos contienen esa variable por año.

Proseguimos creando un nuevo dataframe usando la función melt. Si el nuevo dataframe esta vacio quiere decir que no hay datos para esa fecha y se regres un texto.
En dado caso que el dataframe no este vacio se crea la figura utilizando la biblioteca seaborn y se retorna la figura.

```python
def plotting_siembre_cosecha(year, mes, producto):
  datos_productos = siembra_cosecha_sequia_del_maiz if producto =='Maiz' else (siembra_cosecha_sequia_del_frijol if producto == 'Frijol' else siembra_cosecha_sequia_del_trigo)
  datos_consulta = datos_productos.query(f'Año == {year} and Mes == {mes} and Estado != "Total"')
  datos_consulta = datos_consulta.melt(id_vars=['Estado', 'Tipo_sequia'],value_vars=['Sembrada_ha', 'Cosechada_ha'])

  if datos_consulta.empty:
    return "No hay datos para estas fechas, intente otra"
  else:
    g = sns.catplot(
      data=datos_consulta, kind="bar",
      x="Estado", y="value", hue="variable",
      ci=None, height=5, aspect=3
    )

    ax = g.facet_axis(0,0)
    for p in ax.patches:
        ax.text(p.get_x() + 0.10, 
                p.get_height() * 1.02, 
                '{0:.0f}'.format(p.get_height()), 
                color='black', rotation='horizontal', size='large', ha="left", va = 'bottom')

    g.despine(bottom=True)
    g.set(title="Sembrado y cosechado de "+ producto + " y el tipo de sequia que prensetaba en la fecha: " + str(year) + "-" + str(mes))
    g.set_axis_labels("Estado y Tipo de Sequia que presentaba","Superficie (ha)",fontsize = 20 )
    g.legend.set_title("Variable que se presenta")

    g.fig.set_figwidth(30)
    g.fig.set_figheight(10)

    return g
```

##### Datos sobre la temperatura de México (CONAGUA)

Pero al final tenemos una lista de dataframes, sería preferible tener un único dataframe con todos los datos de todos los años para el análisis estadístico. Para esto, hemos de acomodar nuestros datos de manera que los meses sean una variable "fecha" y estén relacionados a su promedio. De esta forma los análisis serían más sencillos a nivel lista, y en la concatenación todos los dataframes tendrían los mismos índices de columnas

```python
def concatenar(dataframe, nombre): #función de concatenación de los dataframes
  def acomodar(data_frame, i, nombre): #función de acomodar los dataframes con el mes como variable
    return pd.melt(data_frame[i], id_vars= ["ENTIDAD"], value_vars= data_frame[i].columns[1:]).rename(columns = {"variable" : "FECHA", "value" : nombre})
  lista_acomodados = [] #lista de los dataframes acomodados
  for i in range(19): #aplicación de la función a los dataframes
   lista_acomodados.append(acomodar(dataframe, i, nombre)) #guardado de los dataframes a la lista
  dataframe_concatenados = pd.concat(lista_acomodados).reset_index(drop = True) #unión de los dataframes
  return dataframe_concatenados
```

####Estados con condiciones optimas para el sembrado de maíz

Para nuestro análisis es necesario eliminar los datos "NACIONAL", ya que solo estamos interesados en los estados. Ademas de eliminar los resumenes anuales. Y por ultimo agregar 1.5 grados a las temperaturas, que seria la variación de la temperatura suelo-ambiente

```python
Temp_suelo = Temp
Temp_suelo["Temperatura_promedio"] = (Temp_suelo["Temperatura_promedio"] + 1.5)
Temp_suelo["Temperatura_minima"] = (Temp_suelo["Temperatura_minima"] + 1.5)
Temp_suelo["Temperatura_maxima"] = (Temp_suelo["Temperatura_maxima"] + 1.5)
```

```python
Temp_estados = Temp_suelo[(Temp["ENTIDAD"] != "NACIONAL") & (Temp_suelo["Mes"] != "ANUAL")] #datos unicamente de estados por mes
Temp_estados.shape #número de reportes
```

Vemos que contamos con 7104 casos a diferentes años desde 2003 hasta 2021 por mes

Al analizar los limites de temperatura para la siembra de maíz encontramos algo peliculiar

```python
Temp_estados[(Temp_estados["Temperatura_maxima"] > 40)] #estados que superaron el límite de 40°C
```

No muchos casos superan el límmite superior de 40°C, y lo que es mas los casos parecieran aleatorios a primera vista

Pero para el limite inferior es otra historia

```python
Temp_estados[(Temp_estados["Temperatura_minima"] < 10)].shape #casos que superarón el límite de 10°C
```

En todos los registros 1259 veces el límite inferior se supero, por lo que México tiene más problemas con el frio que con el calor, y en el caso del calor, pareciera no haber un cambio, solo un dato muy aleatorio

```python
Temp_estados[(Temp_estados["Temperatura_minima"] > 10 ) & (Temp_estados["Temperatura_maxima"] < 40)].shape #casos dentro del límite 10°C - 40°C
```

Concluyendo con los limites, 5810 veces algunos estados, algunos meses, algunos años, se mantuvieron dentro de los limites para el sembrado de maíz

En cuanto a las condiciones optimas

```python
#Casos optimos
Temp_estados[(Temp_estados["Temperatura_promedio"] < 30) & (Temp_estados["Temperatura_promedio"] > 20) & (Temp_estados["Temperatura_minima"] > 10 ) & (Temp_estados["Temperatura_maxima"] < 40)].shape
```

En 4538 de los casos se cumplieron las condiciones optimas para el sembrado de maíz, las temperaturas se mantuvieron dentro de los limites, y en promedio se encontro dentro del rango optimo 20°C-30°C

En un futuro se busca estudiar a fondo los casos optimos y dentro de los límites, agrupandolos por años, por meses, para determinar si hay un incremento en los casos fuera de los límites. También se busca investigar si el comportamiento es aleatorio, con el fin de indagar si el cambio de temperatura en México es un problema a conciderar a futuro para la siembra de maíz
